# Prepare our data for Modelling/Preprocessing

```python
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()
numerical=['age','education-num','capital-gain','capital-loss','hours-per-week']
df[numerical]=scaler.fit_transform(df[numerical])
```

```python
df
```

```python
y=df['Income']
```

```python
y
```

```python
df.drop('Income',axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

encoder=LabelEncoder()
y=y.apply(lambda x:0 if x=='<=50K' else 1)
```

```python
y
```

```python
X = pd.get_dummies(df)
X
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test= train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
X_train
```

```python
Y_train
```

# Decision Tree classifier

```python
from sklearn.tree import DecisionTreeClassifier

dt=DecisionTreeClassifier(max_depth=11, min_samples_split=90)
dt.fit(X_train,Y_train)
print(dt.score(X_train,Y_train))
```

```python
prediction=dt.predict(X_test)
prediction
```

```python
from sklearn.metrics import accuracy_score
accuracy_dt=accuracy_score(Y_test,prediction)*100
accuracy_dt
```

# Increasing Accuracy by hypertuning the parameters

```python
from sklearn.model_selection import GridSearchCV

parameters={'min_samples_split':range(10,500,20),'max_depth':range(1,20,2),'criterion':['gini','entropy']}
clf_tree=DecisionTreeClassifier()
clf=GridSearchCV(clf_tree,parameters,cv=5)
clf.fit(X_train,Y_train)
```

```python
clf.best_estimator_
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
log.fit(X_train,Y_train)
log
```

```python
log_test=log.predict(X_test)
log_test
```

```python
accuracy_score(log_test,Y_test)
```
